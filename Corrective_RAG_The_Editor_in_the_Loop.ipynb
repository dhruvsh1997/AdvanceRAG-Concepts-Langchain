```python
# Step 1b: Import the essentials

import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough, RunnableLambda, RunnableParallel
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage
from langchain.memory import ConversationBufferMemory
from langchain_core.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field
from pydantic import BaseModel, Field
from google.colab import drive
```

```python
# Step 2: Mount your Google Drive to access the LAWDoc folder
drive.mount('/content/drive')
```

```python
# Step 2b: Load your 15 PDFs using PyPDFLoader
# Assuming your files are in '/content/drive/My Drive/LAWDoc/'
pdf_folder_path = '/content/drive/My Drive/Colab Notebooks/DataProcessing/LAWDoc/'

# Collect all PDF file paths
import glob
pdf_files = glob.glob(pdf_folder_path + '*.pdf')

# Load and split PDFs into Documents
documents = []
for pdf_file in pdf_files:
    loader = PyPDFLoader(pdf_file)
    documents.extend(loader.load())

print(f"Loaded {len(documents)} total document chunks from your 15 PDFs.")
```

```python
# Step 3: Embed with 'avsolatorio/GIST-Embedding-v0'
embedding_model = HuggingFaceEmbeddings(model_name="avsolatorio/GIST-Embedding-v0")

# Create FAISS vectorstore
vectorstore = FAISS.from_documents(documents, embedding_model)

# Save the index for re-use (optional)
vectorstore.save_local("faiss_lawdb")
```

```python
import polars as pl

# Assuming the `documents` list from the previous code block is available
if 'documents' in locals():
    # Extract metadata and content for Polars DataFrame
    data_for_df = []
    for i, doc in enumerate(documents):
        data_for_df.append({
            "chunk_index": i,
            "page_content": doc.page_content,
            "metadata": doc.metadata # This will be a dictionary
        })

    # Create the Polars DataFrame
    df = pl.DataFrame(data_for_df)

    # Unnest the metadata column for better readability
    # This assumes the metadata dictionary keys are consistent across documents
    if not df["metadata"].is_empty():
        df = df.unnest("metadata")

    # Show the DataFrame
    print(df)
else:
    print("The 'documents' list is not available. Cannot create DataFrame from original documents.")
```

```python
# Step 4: Setup Retriever with Maximal Marginal Relevance (MMR) and custom Lambda
retriever = vectorstore.as_retriever(
    search_type="mmr",
    search_kwargs={
        "k": 5,             # Number of docs to retrieve
        "lambda_mult": 0.2  # Balancing diversity and relevance
    }
)

print("Retriever is ready with MMR strategy.")
```

```python
# Step 5: Load your Groq LLMs with your API key (set it in your environment)
from google.colab import userdata
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

# First pass LLM for basic RAG
llm_rag = ChatGroq(model_name="gemma2-9b-it", temperature=0.2)

# Second pass Editor LLM for corrective generation
llm_editor = ChatGroq(model_name="llama-3.3-70b-versatile", temperature=0.3)
```

```python
class CorrectedAnswer(BaseModel):
    answer: str = Field(description="The improved answer, refined with the user's feedback.")

# Note: Use `pydantic_object` instead of `pydantic_schema`
parser = PydanticOutputParser(pydantic_object=CorrectedAnswer)
```

```python
# Step 9: Build the full RAG chain with Editor-in-the-Loop
# Basic RAG: Retrieve & Generate
rag_chain = RunnableParallel({
    "context": retriever,
    "question": RunnablePassthrough()
}) | RunnableLambda(
    lambda input: llm_rag.invoke(f"Context:\n{input['context']}\nQuestion:\n{input['question']}")
)
rag_chain
```

```python
# Step 8: Keep conversation context for multiple passes
memory = ConversationBufferMemory()
```

```python
# Corrective Loop: Take RAG output and user feedback, run through Editor
def corrective_loop(rag_answer, user_feedback):
    format_instructions = parser.get_format_instructions()
    # Construct the prompt message based on whether user_feedback is provided
    if user_feedback:
        editor_prompt_content = f"""
                                  Original Answer: {rag_answer}

                                  Feedback: {user_feedback}

                                  Please provide an improved version based on the feedback, in the specified JSON format.

                                  {format_instructions}
                                  """
    else:
        editor_prompt_content = f"""
                                  Original Answer: {rag_answer}

                                  Please refine and expand this answer, in the specified JSON format.

                                  {format_instructions}
                                  """

    # Step 6: Define the system and human prompt for corrections
    editor_prompt = ChatPromptTemplate.from_messages([
        SystemMessage(
            content="You are a legal editor AI. Your role is to refine, expand, and correct the answer using user feedback or by improving the original answer if no feedback is given. Always respond in the specified JSON format."
        ),
        HumanMessage(
            content=editor_prompt_content
        )
    ])
    response = llm_editor.invoke(editor_prompt.format_messages(user_feedback=user_feedback))

    # DEBUG: see raw output
    print("Raw LLM Editor Response:", response)

    return parser.invoke(response)

# Final orchestrator
def corrective_rag_pipeline(question, user_feedback):
    rag_answer = rag_chain.invoke(question)
    # Ensure rag_answer is treated as a string and stored correctly in memory
    rag_answer_text = rag_answer.content if hasattr(rag_answer, "content") else str(rag_answer)
    memory.save_context({"question": question}, {"answer": rag_answer_text})

    corrected_answer = corrective_loop(rag_answer_text, user_feedback)
    memory.save_context(
        {"user_feedback": user_feedback},
        {"corrected_answer": corrected_answer.answer}  # Use field, not whole object
    )
    return corrected_answer

print("Corrective RAG pipeline is ready for questions.")
```

```python
# Step 10: Try it out!
# Example question
question = "How do the limitations of the 'commercial optical non-contact profiler' and the 'transmission Mach–Zehnder interferometer' collectively impact the comprehensive characterization of both geometrical and optical parameters, especially when considering real-world fabrication imperfections?"
# Simulated user feedback
user_feedback = "Please update the answer since the question I am asking for is related to 'different microlens fabrication techniques and the methods used to characterize their properties' Domain."

corrected_response = corrective_rag_pipeline(question, user_feedback)
print("Corrected Answer:", corrected_response)
```
